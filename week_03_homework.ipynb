6. What is the correlation coefficient of `mpg` and `weight`?

```python
df1 = pd.DataFrame(mpg)
df2 = pd.DataFrame(weight)
correlation_mpg = df1.corr(method='pearson')
print(correlation_mpg)
correlation_weight = df2.corr(method='pearson')
print(correlation_weight)
```
